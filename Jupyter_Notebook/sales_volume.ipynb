```python
def volume_trend(row,cols):
#     print()
    
#     print()
    averages = row[cols].values.tolist()
    min_range = row[cols].mean()-5
    max_range = row[cols].mean()+5
    res= all(ele >= min_range and ele < max_range for ele in row[cols])
    if(averages == sorted(averages))and  averages.count(0)<2 and len(set(averages))!= 1 and len(set(averages))== len(averages):
        return "increasing trend"
    elif(averages == sorted(averages,reverse=True)) and averages.count(0)<2 and len(set(averages))!= 1 and len(set(averages))== len(averages):

        return "declining trend"
    elif res==True:

        return "flat trend"
    else:

        return "fluctuating trend"
#     print(averages)
```

```python
cols = df_prd.drop(columns = ['customer','product']).columns

df_prd['segments'] = df_prd.iloc[:,2:].apply(lambda row: volume_trend(row,cols),axis = 1)
cols
```

```python
df_prd1 = df_prd.groupby(['segments']).sum().reset_index()
df_prd2 = df_prd.groupby(['segments']).count().reset_index().drop(columns=cols).drop(columns='product')
```
